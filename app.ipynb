# Вводимо OpenAI API ключ

```python
api_key = input("Введіть OpenAI API ключ")
```

# Готуємо RAG послідовність

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.runnables import RunnablePassthrough
from langchain_core.output_parsers import StrOutputParser
from tools.util import create_vector_db_retriever, split_docs
from tools.llm import get_llm

llm = get_llm(api_key)
retriever = create_vector_db_retriever(split_docs(), api_key)

template = """Ти - фінансовий радник, котрий може відповідати на питання про фінансові показники. Не відповідай на питання, котрі не стосуються фінансових та кадрових питань певної компанії.

Контекст: {context}

Питання: {question}
"""

prompt = ChatPromptTemplate.from_template(template)
```

# Створюємо послідовність RAG

```python
from langchain_core.runnables import RunnableParallel
from tools.util import format_docs

rag_chain = (
    RunnableParallel({"context": retriever | format_docs, "question": RunnablePassthrough()})
    | prompt
    | llm
    | StrOutputParser()
)
```

# RAG відповідає на ваші питання

```python
print(rag_chain.invoke(input("Запитуйте фінансового консультанта")))
```
